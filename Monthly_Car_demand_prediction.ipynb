```
import pandas as pd
# read the files
path='/content/drive/My Drive/Motor Vehicle Prediction/'
df=pd.read_csv(path+'Merged File.csv')
```

```
df.head()
```

```
df=df.interpolate(method='linear', order=2,limit_direction ='forward')
df=df.interpolate(method='linear', order=2,limit_direction ='backward')
```

```
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X=df.iloc[:,18:268]
Y=df.iloc[:,1]
bestfeatures=SelectKBest(score_func=chi2, k=40)
fit=bestfeatures.fit(X,Y)
dfscores=pd.DataFrame(fit.scores_)
dfcolumns=pd.DataFrame(X.columns)
featuresScores=pd.concat([dfcolumns,dfscores],axis=1)
featuresScores.columns=['feature','score']
print(featuresScores.nlargest(40,'score'))
```

```
featuresScores=featuresScores.sort_values(by='score',ascending=False)
importantFeatures=featuresScores.feature.values.tolist()
importantFeatures=importantFeatures[0:40]
X=X=pd.DataFrame(X,columns=importantFeatures)
```
